```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from cal import get_molecular_features
import cal
from sklearn.model_selection import train_test_split as tts
from rdkit import Chem
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
df.columns
```

```python

aroma=[cal.calculate_aromatic_proportion(Chem.MolFromSmiles(smile)) for smile in df['SMILES']]
```

```python
df['aromatic propotion']=aroma
```

```python
features = [
    'MolWt', 'MolLogP', 'MolMR', 'HeavyAtomCount', 
    'NumHAcceptors', 'NumHDonors', 'NumHeteroatoms', 
    'NumRotatableBonds', 'NumValenceElectrons', 'NumAromaticRings', 
    'RingCount', 'TPSA', 'LabuteASA', 'aromatic propotion','Solubility'
]
df1=df[features]
```

```python
df1
```

```python
X,y=df1.drop(columns='Solubility'),df1['Solubility']
df1
```

```python
df1
```

```python
rf=RandomForestRegressor(max_depth=40, max_features='sqrt', n_estimators=200,
                      random_state=42)
```

```python
x_train,x_test,y_train,y_test=tts(X,y,train_size=0.8,random_state=42)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error
```

```python
r2_score(y_test,y_pred)
root_mean_squared_error(y_test,y_pred)
```

```python
y_test_pred=rf.predict(x_train)
```

```python
r2_score(y_test_pred,y_train)
```

```python
from sklearn.metrics import r2_score

train_r2 = r2_score(y_train, rf.predict(x_train))
test_r2 = r2_score(y_test, y_pred)

print(f"Training R2: {train_r2:.4f}")
print(f"Testing R2:  {test_r2:.4f}")
```
